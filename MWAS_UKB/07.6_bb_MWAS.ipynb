This notebook performs MWAS using both imputed blood methylome & imputed brain methylome: *phecode (j) ~ covar + blood_DNAm (i) + brain_DNAm (i)*

## Methylome-wide associations

```python
### inputs needed: 
## passed by -iin: batch*_predict.txt (brain based)
## passed by -iin: batch*_predict_blood.txt (blood based)
## passed by -iin: icd.phe 
## passed by -iin: cov.phe
## passed by -iin: phecode_map.csv
## passed by -iin: mwas_func.py
## passed by -iin: mwas_07.6_bb_MWAS.py
## passed by -iin: mwas_07.6_bb_MWAS.sh
## passed by -icmd: $1 batch
```

```python
%%bash

### rename blood-based predictors so they don't have the same name as the brain-based

for i in {1..50}; do
    file_in="vasilis/data/ebb/scores_blood/batch${i}_predict.txt" 
    file_out="vasilis/data/ebb/scores_blood/batch${i}_predict_blood.txt"
    dx mv $file_in $file_out
done
```

```python
%%bash

icd="vasilis/data/ebb/phenotypes/icd.phe"
cov="vasilis/data/ebb/phenotypes/cov.phe"
phemap="vasilis/data/ebb/phenotypes/phecode_map.csv"
predictdir_blood="vasilis/data/ebb/scores_blood"
predictdir_brain="vasilis/data/ebb/scores"
dest="vasilis/data/ebb/results_bb/"

for batch in {2..50}; do
    dx run swiss-army-knife \
        -iin="vasilis/SAK_scripts/mwas_07.6_bb_MWAS.sh" \
        -iin="vasilis/SAK_scripts/mwas_07.6_bb_MWAS.py" \
        -iin="vasilis/SAK_scripts/mwas_func.py" \
        -iin="${predictdir_blood}/batch${batch}_predict_blood.txt" \
        -iin="${predictdir_brain}/batch${batch}_predict.txt" \
        -iin="${icd}" \
        -iin="${cov}" \
        -iin="${phemap}" \
        -icmd="sh mwas_07.6_bb_MWAS.sh ${batch}" \
        --tag="as_${batch}" \
        --instance-type "mem1_ssd1_v2_x36" \
        --destination="${dest}" \
        --brief --yes --priority high
done
```
